Scratch implementation

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Load digits dataset
digits = load_digits()
X, y = digits.data, digits.target
print("X shape:", X.shape)   # (n_samples, n_features)
print("y shape:", y.shape)   # (n_samples,)
  

# Standardize the data (important for PCA)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
print("Feature means (first 5):", np.mean(X_scaled, axis=0)[:5])
print("Feature stds  (first 5):", np.std(X_scaled, axis=0)[:5])

```

```python
# PCA implementation from scratch

def compute_pca(X, n_components):
    # Step 1: Center the data (subtract mean)
    X_meaned = X - np.mean(X, axis=0)

    # Step 2: Compute covariance matrix
    cov_matrix = np.cov(X_meaned.T)

    # Step 3: Compute eigenvalues and eigenvectors
    eigenvalues, eigenvectors = np.linalg.eigh(cov_matrix)

    # Step 4: Sort eigenvalues (descending) and reorder eigenvectors
    sorted_indices = np.argsort(eigenvalues)[::-1]
    eigenvalues = eigenvalues[sorted_indices]
    eigenvectors = eigenvectors[:, sorted_indices]

    # Step 5: Select top n_components eigenvectors
    eigenvectors_subset = eigenvectors[:, :n_components]

    # Step 6: Project data onto principal components
    X_reduced = np.dot(X_meaned, eigenvectors_subset)

    # Step 7: Return reduced data and eigen info
    return X_reduced, eigenvalues, eigenvectors_subset
```

```python


# Reduce to 30 principal components
n_components = 30
X_pca, components, eigenvalues = compute_pca(X_scaled, n_components)

# Split data into train/test
X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=42)

# Train a classifier (Logistic Regression)
clf = LogisticRegression(max_iter=1000)
clf.fit(X_train, y_train)

# Predict and evaluate
y_pred = clf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Classification accuracy after PCA: {accuracy:.4f}")

# Optional: plot the first 2 principal components
plt.figure(figsize=(8,6))
for digit in np.unique(y):
    plt.scatter(X_pca[y==digit, 0], X_pca[y==digit, 1], label=str(digit), alpha=0.6)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('Digits Dataset Projected onto First 2 Principal Components')
plt.legend()
plt.show()
```

We get less accuancy (~60%) as using only 2 components did not retain much of the feature information. However in real life you will find many cases where using 2 or few PCA components can still give you a pretty good accuracy

Download heart disease dataset heart.csv in Exercise folder and do following, (credits of dataset: https://www.kaggle.com/fedesoriano/heart-failure-prediction)

Load heart disease dataset in pandas dataframe
Remove outliers using Z score. Usual guideline is to remove anything that has Z score > 3 formula or Z score < -3
Convert text columns to numbers using label encoding and one hot encoding
Apply scaling
Build a classification model using various methods (SVM, logistic regression, random forest) and check which model gives you the best accuracy
Now use PCA to reduce dimensions, retrain your model and see what impact it has on your model in terms of accuracy. Keep in mind that many times doing PCA reduces the accuracy but computation is much lighter and that's the trade off you need to consider while building models in real life

```python


# Step 1: Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.decomposition import PCA
from scipy import stats

# Step 2: Load dataset
df = pd.read_csv("heart - heart.csv")
print("✅ Dataset loaded successfully!")
print(df.head())

# Step 3: Handle missing values
df = df.dropna()

# Step 4: Remove outliers using Z-score
numeric_df = df.select_dtypes(include=[np.number])
z_scores = np.abs(stats.zscore(numeric_df))
df = df[(z_scores < 3).all(axis=1)]
print(f"✅ After removing outliers, shape = {df.shape}")

# Step 5: Encode categorical columns
cat_cols = df.select_dtypes(exclude=[np.number]).columns
if len(cat_cols) > 0:
    le = LabelEncoder()
    for col in cat_cols:
        df[col] = le.fit_transform(df[col])
    print(f"Encoded categorical columns: {list(cat_cols)}")
else:
    print("No categorical columns detected.")

# Step 6: Split features and target
target_col = 'target' if 'target' in df.columns else df.columns[-1]
X = df.drop(target_col, axis=1)
y = df[target_col]

# Step 7: Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 8: Train/Test split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

# Step 9: Train models (without PCA)
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "SVM": SVC(kernel='rbf'),
    "Random Forest": RandomForestClassifier(n_estimators=100, random_state=42)
}

acc_no_pca = {}

print("\n🔹 Model Performance Without PCA 🔹")
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    acc_no_pca[name] = acc
    print(f"{name} Accuracy: {acc:.4f}")

# Step 10: Apply PCA
pca = PCA(n_components=5)
X_pca = pca.fit_transform(X_scaled)

# Split after PCA
X_train_pca, X_test_pca, y_train_pca, y_test_pca = train_test_split(
    X_pca, y, test_size=0.2, random_state=42
)

# Step 11: Train models (with PCA)
acc_with_pca = {}

print("\n🔹 Model Performance With PCA 🔹")
for name, model in models.items():
    model.fit(X_train_pca, y_train_pca)
    y_pred = model.predict(X_test_pca)
    acc = accuracy_score(y_test_pca, y_pred)
    acc_with_pca[name] = acc
    print(f"{name} Accuracy (with PCA): {acc:.4f}")

# Step 12: Visualization — Compare accuracies
plt.figure(figsize=(10, 6))
bar_width = 0.35
index = np.arange(len(models))

plt.bar(index, list(acc_no_pca.values()), bar_width, label='Without PCA', color='skyblue')
plt.bar(index + bar_width, list(acc_with_pca.values()), bar_width, label='With PCA', color='lightgreen')

plt.xlabel('Models', fontsize=12)
plt.ylabel('Accuracy', fontsize=12)
plt.title('Model Accuracy Comparison: Before vs After PCA', fontsize=14)
plt.xticks(index + bar_width / 2, list(models.keys()))
plt.legend()
plt.tight_layout()
plt.show()

```
